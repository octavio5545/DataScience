```python
def filtro_repetidos(df, drop_colName, group_colName):
    df_filtrado = df.drop_duplicates(drop_colName)\
    .groupby(by=[group_colName])\
    .filter(lambda f1: f1.shape[0]>1) \
    .sort_values(group_colName, ascending=True)
    
    return df_filtrado
```
