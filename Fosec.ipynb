```python
#DEVUELVE DATAFRAME CON MÍNIMOS Y MÁXIMOS DE UN CAMPO.

def min_max_subgroups(df, cols, key, names_groups=None, save=False):
    #import pandas as pd
    
    # cols: Son las columnas del df que se quieren analizar
    # key : Nombre de la columna de agrupación
    # names_groups: lista de dos entradas con los nombres de los dataframes
    #               con los mínimos y máximos respectivamente.
    # save: True si se quiere generar los csv con mínimos y máximos.
    
    grouped_df = df[cols].groupby(cols)\
    .size().reset_index(name="Count")
    
    min_group = grouped_df.groupby([key]).min()
    max_group = grouped_df.groupby([key]).max()
    if save:
        min_group.to_csv(names_groups[0])
        max_group.to_csv(names_groups[1])
    
    return min_group, max_group
```

```python
fond_conv_min, fond_conv_max = min_max_subgroups(df, cols=['CLAVE_FONDO', 'CVE_CONVOCATORIA'], key='CLAVE_FONDO', 
                                                     names_groups=["Min_conv_por_fondo_FOMIX.csv", "Max_conv_por_fondo_FOMIX.csv"],
                                                     save=True)
```

### Número máximo y mínimo de convocatorias en los fondos:

```python
fondos_y_conv = df[['CLAVE_FONDO', 'CVE_CONVOCATORIA']]
conv_unicas_por_fondo = fondos_y_conv.groupby('CLAVE_FONDO').nunique()
max_conv = int(conv_unicas_por_fondo.max())
min_conv = int(conv_unicas_por_fondo.min())
```

```python
#FONDOS QUE TIENEN EL MÍNIMO DE CONVOCATORIAS ÚNICAS:
conv_unicas_por_fondo[conv_unicas_por_fondo['CVE_CONVOCATORIA'] == min_conv]
```

```python
#FONDOS QUE TIENEN EL MÁXIMO DE CONVOCATORIAS ÚNICAS:
conv_unicas_por_fondo[conv_unicas_por_fondo['CVE_CONVOCATORIA'] == max_conv]
```
